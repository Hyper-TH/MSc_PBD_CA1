```python
# Import the necessary pyspark functions
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.ml.feature import StringIndexer, OneHotEncoder
```

```python
# Set up Session
from pyspark.sql import SparkSession

spark = SparkSession.builder \
        .appName('C20361521 - Assignment 1') \
        .master('local[*]') \
        .getOrCreate()
```

```python
# Load both datasets into DFs
train_file_path = 'datasets/train.csv'
test_file_path = 'datasets/test.csv'

train_df = spark.read.format("csv").option("header", "true")\
    .option("inferSchema", "true")\
    .load(train_file_path)

test_df = spark.read.format("csv").option("header", "true")\
    .option("inferSchema", "true")\
    .load(test_file_path)
```

```python
# Cache it with a count operation
print('Training set size = ', train_df.cache().count())
print('Training set size = ', test_df.cache().count())

# train_df.show(5)
```

```python
# Check schema
train_df.printSchema()
```

```python
test_df.printSchema()
```

## Data Exploration

```python
# Will use TempView to explore and analyse the dataframes
# Note to self: refer to this https://spark.apache.org/docs/latest/sql-ref-syntax-qry-select.html

train_df.createOrReplaceTempView("train_tab")
test_df.createOrReplaceTempView("test_tab")
```

```python
# Check all possible ages in the training data
spark.sql("""SELECT distinct(`Age at enrollment`) as AGE FROM train_tab ORDER BY AGE""").show(train_df.count())
```

```python
# Check all possible ages in the testing data
spark.sql("""SELECT distinct(`Age at enrollment`) as AGE FROM train_tab ORDER BY AGE""").show(test_df.count())
```

```python
# Also check the summary of the Age at enrollment column
train_df.select("Age at enrollment").summary().show()
test_df.select("Age at enrollment").summary().show()

## Same quarters, min and max
```

```python
# Distribution of the Admission Grade
train_df.select("Admission grade").summary().show()
```

```python
# Distribution of Unemployment Rate
train_df.select("Unemployment rate").summary().show()
```

```python
# Scholarship status
train_df.select(when(col("Scholarship holder") == 0, "No").otherwise("Yes").alias("Scholarship_Status")).groupBy("Scholarship_Status").count().show()
```
