```python
# Find the most common finished education per age
spark.sql("""
    SELECT `Previous Qualification` as Most_Common_Qualification, `Age at enrollment` AS Age
        FROM (
            SELECT `Previous Qualification`, `Age at enrollment`,
                RANK() OVER (PARTITION BY `Age at enrollment` ORDER BY cnt DESC) as rnk 
            FROM (
                SELECT `Previous Qualification`, `Age at enrollment`, COUNT(*) AS cnt
                FROM train_tab 
                GROUP BY `Previous Qualification`, `Age at enrollment`
            ) AS qual_count
        ) AS qual_rank_per_age
        WHERE rnk = 1
    """).show(train_df.count())
```

```python
# Find the most common coures per gender
spark.sql("""
    SELECT `Course` as Most_Common_Course, `Gender`
        FROM (
            SELECT `Course`, `Gender`,
                RANK() OVER (PARTITION BY `Gender` ORDER BY cnt DESC) as rnk 
            FROM (
                SELECT `Course`, `Gender`, COUNT(*) AS cnt
                FROM train_tab 
                GROUP BY `Course`, `Gender`
            ) AS course_count
        ) AS course_rank_per_age
        WHERE rnk = 1
    """).show(train_df.count())

# 0 (Female) => 9500 Nursing
# 1 (Male) => 9147 Management 
```

### Convert Categorical Variables

```python
# Sample code
categoricalCols = ["workclass", "education", "marital_status", "occupation", "relationship", "race", "sex"]

# The ff two lines are estimators. Return funcions that we'll later apply to transform the dataset.
stringIndexer = StringIndexer(inputCols=categoricalCols, outputCols=[x + "Index" for x in categoricalCols], handleInvalid="keep")
encoder = OneHotEncoder(inputCols=stringIndexer.getOutputCols(), outputCols=[x + "OHE" for x in categoricalCols])

# The label column 'income' is also a string value, has two possible values "<=50K" and ">50K".
# Convert it to a numeric value (perhaps this one is the Target?)
labelToIndex = StringIndexer(inputCol="income", outputCol="label")
```
